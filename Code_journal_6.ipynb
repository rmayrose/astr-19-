Import the *Math* module to work with sine and cosine.

```python
import math
```

Create a python function that returns *sin(x)*.

```python
def my_sin(x):
    return math.sin(x)
```

Create a python function that returns *cos(x)*.

```python
def my_cos(x):
    return math.cos(x)
```

Tabulate x between values of 0 and 2pi with 1000 entries.

```python
import math

num_points = 1000
x_min = 0.0
x_max = 2 * math.pi
step = (x_max - x_min) / (num_points - 1)

print("      x          sin(x)            cos(x)")
print("------------------------------------------------")

for i in range(num_points):
    x = x_min + i * step
    s = my_sin(x)
    c = my_cos(x)
    print(f"{x:12.6f}  {s:12.6f}  {c:12.6f}")
```
