```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,1,100)
x
```

```python
def my_sin(x):
    return np.sin(x)
def my_cos(x):
    return np.cos(x)
```

```python
import math
x_sin = my_sin(x)
x_cos = my_cos(x)

fig, axes = plt.subplots(1,2)
axes[0].plot(x_sin,label="sin(x)")
axes[0].plot(x)
axes[0].legend()

axes[1].plot(x_cos,label="cos(x)")
axes[1].plot(x)
axes[1].legend()

plt.savefig("journal8_fig")
plt.show()
```
