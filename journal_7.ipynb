```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,1,100)
x
```

```python
import math
def my_exp(x):
    return math.exp(x)
```

```python
y = np.array([my_exp(val) for val in x])
plt.plot(x,y)
plt.xlabel("Time [milliseconds]")
plt.ylabel("Awesomeness")

plt.savefig('Awesomeness')
```
